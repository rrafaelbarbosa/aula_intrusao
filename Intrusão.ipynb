```python
# em geral um banco de dados  NÃO ESTÁ NO PYTHON
# um banco de dados é conectado a um backend
# é comum ser em SQL - Firebase - Supabase - Postgree....

# vamos criar nossa base de dados vazia
#  { } < dicionario que é multavel ( ) < tupla que é imutável
usuarios = {}
# função de cadastro e login
def cadastrar():
  novo_usuario = input('informe o nome do usuario')
  nova_senha = input('informe a senha')

  # para adicionar um valor no dicionario
  # dicionario[chave] = valor
  usuarios[novo_usuario] = nova_senha
  print('usuario cadastrado com sucesso')

import getpass
def login():
  usuario = input('informe seu usário')
  senha = getpass.getpass('informe sua senha')

  # vamos verificar se o usuario digitado está no dicionario usuarios
  # e a senha informada é o valor da chave
  if usuario in usuarios and usuarios[usuario] == senha:
    print('login realizado com sucesso')
  else:
    print('usuario e/ou senha incorretos')

# executar o sistema
while True:
  # while True executa um loop sem condição e é finalizado com o break
  print('1 - cadastrar novo usuario')
  print('2 - ver usuários cadastrados')
  print('3 - ver senhas cadastradas')
  print('4 - ver usuarios e senhas')
  print('5 - logar')
  print('0 - sair do sistema')

  op = input('informe sua escolha')

  if op == '1':
    cadastrar()
  elif op == '2':
    # a função dicionario.keys() - exibe todas as chaves cadastradas
    print(usuarios.keys())
  elif op == '3':
    #a função dicionario.values() - exibe todos os valores cadastrados
    print(usuarios.values())
    # a função dicionario.items() - exibe todas as combinações cadastradas
  elif op == '4':
    print(usuarios.items())
  elif op == '5':
    login()
  elif op == '0':
    break
  else:
    print('opção não encontrada')
```
